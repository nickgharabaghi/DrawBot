```python
# First: loading an image of a maze into Python

import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# RULES FOR MAZE IMAGE:
# - Outer border of image must be the walls, and it must be a dark colour
# - The start of the maze must be in the top row, and the end in the bottom. 
# - The start and end of the maze must be empty spaces in the top and bottom rows.

# Load in image:
filename_base = r'C:\Users\nickg\Desktop\Side Projects\Test images\\' # Need a double backslash here to work properly with the apostrophe
filename = 'maze2.jpg' 
image1_colour = mpimg.imread(filename_base + filename)
# Show the inputted image, to make sure it's what you want:
print("Inputted image:")
image1plot = plt.imshow(image1_colour) 
plt.show()

# Notes about working with a maze image in a .jpg format:
# In color, [0, 0, 0] is black, [255, 255, 255] is white. 
# Due to resolution I assume, there is some blurring around the lines of the maze. You'll get things
# like [35, 35, 35] at the pixels at the black-white transition.

# I'm going to try to convert images to grayscale by taking the "grayscale pixel value" to be the
# average of the R, G, and B pixel values in the original image. For a black and white image, this shouldn't
# really be too different from just using the first index, but might as well try to future proof for colour
# images while I can.

# This could absolutely stand to be optimized.
image1_gray = np.zeros([image1_colour.shape[0], image1_colour.shape[1]])
for i in range(0, image1_gray.shape[0]):
    for j in range(0, image1_gray.shape[1]):
        image1_gray[i,j] = np.average(image1_colour[i,j,:])
print('Grayscale version of inputted image:')
image1gray = plt.imshow(image1_gray, cmap = 'gray') 
plt.show() # Show the grayscale version


# Next, would like some way for the program to identify where the empty spaces are, and connect them somehow.
# This represents the parts of the maze that are not walls.

empty_space_val = 40
wall_val = 0
pixel_threshold = 140 # any pixel with a value below this is a wall, and above this is a space

# Create a version of the maze with only black and white:
maze_raw = np.zeros(image1_gray.shape)
for i in range(0, maze_raw.shape[0]):
    for j in range(0, maze_raw.shape[1]):
        if image1_gray[i,j] > pixel_threshold:
            maze_raw[i,j] = empty_space_val # this is empty space
        else:
            maze_raw[i,j] = wall_val # this is a wall
print('All pixels set to one of two values')
image1raw = plt.imshow(maze_raw, cmap = 'gray') 
plt.show()

# Next, I need to deal with the empty spaces somehow, to turn them into a path instead of a space.
# The first option I thought of is running through the maze over and over, each time turning more pixels
# into walls until there is only one connected path of space. This sounds VERY inefficient unfortunately.
# The second option I thought of was to follow the "hug the left wall" strategy, which to my knowledge should always work eventually

# Step one: let's find the start and end of the maze.
# I'm going to call the top row the "start", and the bottom row the "end"
path_val = 220
end_val = 221
maze_for_testing = np.copy(maze_raw)

(num_rows, num_cols) = maze_for_testing.shape

# Now that I can find the start and end, need to plot a path between them.
# The "keep one hand on the left wall" strategy should be fine
start_col = 0 # The column index for the start of the maze in the first row. The first pixel from L to R that's not a wall.
for i in range(0, num_cols): # go through every pixel in the first row to find the start:
    if maze_for_testing[0, i] != 0:
        start_col = i
        maze_for_testing[0, i] = path_val
        break

# AFTER SOME THINKING, IT MIGHT ACTUALLY BE EASIER TO JUST TRACE THE *WALL*, NOT THE
# EMPTY SPACES ALONG THE WALL. THE PATH WILL JUST BE ONE PIXEL BESIDE THE WALL IN 
# THAT CASE

# I should now have a starting point for the path. Do the same for (all of) the end points:
for i in range(0, num_cols): # go through every pixel in the row to find the end:
    if maze_for_testing[-1, i] != 0:
        maze_for_testing[-1, i] = end_val
# Now, one starting pixel should be colored in with path_val and ALL end pixels should be coloured in with end_val

# Next, create the path:
# - Should need a while true loop, since the length of the maze in pixels isn't knowable beforehand.


cols_visited = [start_col] # Start the path at the [0]th row, [start_col]th column
rows_visited = [0] # Start at the first row

# Try down first (relative to grid)
if maze_for_testing[1,start_col] != wall_val:
    cols_visited.append(start_col)
    rows_visited.append(1)
# Then left
elif maze_for_testing[0, start_col - 1] != wall_val:
    cols_visited.append(start_col - 1)
    rows_visited.append(0)
# Then right
elif maze_for_testing[0, start_col + 1] != wall_val:
    cols_visited.append(start_col + 1)
    rows_visited.append(0)

print(cols_visited, rows_visited)


while True:
    # Order to check in to follow right hand on wall strategy:
    # 1) To the right
    # 2) Straight
    # 3) Left
    # 4) Going to assume you can't go back
    
    def check_down(current_row, current_col):
        next_row = current_row + 1
        next_col = current_col
        
        if maze_for_testing[next_row,next_col] == end_val:
            return -1
        if maze_for_testing[next_row,next_col] == empty_space_val:
            return True
        return False
        
    def check_up(current_row, current_col):
        next_row = current_row - 1
        next_col = current_col
        
        if maze_for_testing[next_row,next_col] == end_val:
            return -1
        if maze_for_testing[next_row,next_col] == empty_space_val:
            return True
        return False
        
    def check_right(current_row, current_col):
        next_row = current_row
        next_col = current_col + 1
        
        if maze_for_testing[next_row,next_col] == end_val:
            return -1
        if maze_for_testing[next_row,next_col] == empty_space_val:
            return True
        return False
        
    def check_left(current_row, current_col):
        next_row = current_row
        next_col = current_col - 1
        
        if maze_for_testing[next_row,next_col] == end_val:
            return -1
        if maze_for_testing[next_row,next_col] == empty_space_val:
            return True
        return False
    
    prev_row = rows_visited[-2]
    prev_col = cols_visited[-2]
    
    current_row = rows_visited[-1]
    current_col = cols_visited[-1]
    
    # Figure out direction we went in:
    direction = None
    if current_row == prev_row and current_col > prev_col:
        # Heading right
        direction = 'r'
    if current_row == prev_row and current_col < prev_col:
        # Heading left
        direction = 'l'
    if current_row > prev_row and current_col == prev_col:
        # Heading down
        direction = 'd'
    if current_row < prev_row and current_col == prev_col:
        # Heading up
        direction = 'u'
    
    go_up = False
    go_down = False
    go_left = False
    go_right = False
    
    if direction == 'r':
        found_direction = check_down(current_row, current_col)
        if found_direction == -1:
            break
        if found_direction:
            go_down = True
            found_direction = True
        else:
            found_direction = check_right(current_row, current_col)
            if found_direction == -1:
                break
            if found_direction:
                go_right = True
                found_direction = True
            else:
                found_direction = check_up(current_row, current_col)
                if found_direction == -1:
                    break
                if found_direction:
                    go_up = True
                    found_direction = True

    if direction == 'u':
        found_direction = check_right(current_row, current_col)
        if found_direction == -1:
                break
        if found_direction:
            go_right = True
            found_direction = True
        else:
            found_direction = check_up(current_row, current_col)
            if found_direction == -1:
                break
            if found_direction:
                go_up = True
                found_direction = True
            else:
                found_direction = check_left(current_row, current_col)
                if found_direction == -1:
                    break
                if found_direction:
                    go_left = True
                    found_direction = True
            
    if direction == 'l':
        found_direction = check_up(current_row, current_col)
        if found_direction == -1:
                break
        if found_direction:
            go_up = True
            found_direction = True
        else:
            found_direction = check_left(current_row, current_col)
            if found_direction == -1:
                break
            if found_direction:
                go_left = True
                found_direction = True
            else:
                found_direction = check_down(current_row, current_col)
                if found_direction == -1:
                    break
                if found_direction:
                    go_down = True
                    found_direction = True

    if direction == 'd':
        found_direction = check_left(current_row, current_col)
        if found_direction == -1:
                break
        if found_direction:
            go_left = True
            found_direction = True
        else:
            found_direction = check_down(current_row, current_col)
            if found_direction == -1:
                break
            if found_direction:
                go_down = True
                found_direction = True
            else:
                found_direction = check_right(current_row, current_col)
                if found_direction == -1:
                    break
                if found_direction:
                    go_right = True
                    found_direction = True
                    
    print('Direction: ', direction)
    if go_down:
        print('down')
    if go_up:
        print('up')
    if go_right:
        print('right')
    if go_left:
        print('left')
    
    # Now, one of the four directions will have been selected. Just move in that direction:
    if go_down:
        cols_visited.append(cols_visited[-1])
        rows_visited.append(rows_visited[-1] + 1)
        maze_for_testing[rows_visited[-1] + 1,cols_visited[-1]] = path_val
    elif go_up:
        cols_visited.append(cols_visited[-1])
        rows_visited.append(rows_visited[-1] - 1)
        maze_for_testing[rows_visited[-1] - 1,cols_visited[-1]] = path_val
    elif go_left:
        cols_visited.append(cols_visited[-1] - 1)
        rows_visited.append(rows_visited[-1])
        maze_for_testing[rows_visited[-1],cols_visited[-1] - 1] = path_val
    elif go_right:
        cols_visited.append(cols_visited[-1] + 1)
        rows_visited.append(rows_visited[-1])
        maze_for_testing[rows_visited[-1],cols_visited[-1] + 1] = path_val
    print(rows_visited)
    print(cols_visited)
        
mazefortesting = plt.imshow(maze_for_testing) 
plt.show()        

# while True:
#     # We base things around current_index
#     current_row = rows_visited[-1]
#     current_col = cols_visited[-1]
#     current_pix = maze_for_testing[current_row, current_col]
#     pixels_finished = len(cols_visited)
    
#     # There are 4 neighbouring pixels to check:
#     # 1 - Above
#     if current_row != 0:
#         pix_up = maze_for_testing[current_row - 1, current_col]
#     #2 - Below:
#     if current_row != num_rows - 1:
#         pix_down = maze_for_testing[current_row + 1, current_col]
#     #3 - Right:
#     if current_col != num_cols - 1:
#         pix_right = maze_for_testing[current_row, current_col + 1]
#     #4 - Left:
#     if current_col != 0:
#         pix_right = maze_for_testing[current_row, current_col - 1]
        
#     # Now, these (up to) 4 neighbouring pixels can be one of 3 things: wall, empty space, or path
#     # The starting pixel only has 3 (or 2) neighbours.
#     # Use conditions to eliminate potential moves until only one is left.
#     go_up = True
#     go_down = True
#     go_left = True
#     go_right = True
    
#     # Geometrical constraints:
#     if current_row = 0:
#         go_up = False # can't go up if there is no up
#     if current_row = num_rows -1:
#         go_down = False # can't go down if there is no down
#     if current_col = 0:
#         go_left = 0 # can't go left if there is no left
#     if current_col = num_cols - 1:
#         go_right = 0 # can't go right if there is no right
    
#     # Walls and path:
#     if go_up:
#         if pix_up == wall_val:
#             go_up = False
#     if go_down:
#         if pix_down == wall_val:
#             go_down = False
#     if go_left:
#         if pix_left == wall_val:
#             go_left = False
#     if go_right:
#         if pix_right == wall_val:
#             go_right = False
#     break
    




# # Check in this order:
# # Right of direction it’s moving
# # In direction it’s moving
# # Left of direction it’s moving
# # Back

        
```
